```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
data.head()
```

```python
print(data.shape)  # print the size of dataframe
print(" ")
print(data.nunique())   # prints the total number of unique data for each column
```

```python
data.groupby("name")
```

```python
# let's find relation between year and selling price
year = np.array(data['year'])
sp = np.array(data['selling_price'])

plt.title("Year v/s Selling Price of Cars", fontsize= 20, color='red')
plt.xlabel("Year",fontsize=16,color='blue')
plt.ylabel("Selling Price",fontsize=16,color='blue')
plt.scatter(year,sp,alpha =.4)
plt.show()
```

```python
# let's now find linear regression line
regression = LinearRegression()
# convert year and sp to a dataframe object
x = DataFrame(data,columns=['year'])
y = DataFrame(data,columns=['selling_price'])
regression.fit(x,y)

regression.predict(x)

plt.scatter(x,y,alpha =.4)
plt.plot(year,regression.predict(x),color='red')
plt.xlim(2003,2020)
plt.ylim(0,4e6)
plt.grid()

plt.title("Year v/s Selling Price of Cars", fontsize= 20, color='red')
plt.xlabel("Year",fontsize=16,color='blue')
plt.ylabel("Selling Price",fontsize=16,color='blue')
plt.show()
```

### Key points to keep in mind

- regression.fit(x,y) takes x and y as a dataframe object
- plt.plot(x,y) can plot with either both series object or both array object but if dataframe object both must have only one series selected
