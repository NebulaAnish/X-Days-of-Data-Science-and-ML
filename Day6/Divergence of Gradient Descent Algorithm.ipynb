# Divergence of Gradient Descent Algorithm on Cost Function

## Cost function is

- $ h(x) = x^5 -2x^4 +2 $

```python
# import section
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Function definiton section
def h(x):
    return x**5-2*x**4 +2
def dh(x):
    return 5*x**4 -8*x**3
```

```python
# sample points
x_3 = np.linspace(-2.5,2.5,1000)
```

```python
# Gradient Descent function
def gradient_descent(derivative_function, first_guess, step_multiplier=.01,precision=.0001):
    new_x = first_guess
    prev_x = 0
    xlist = []
    dxlist = []
    while(1==1):
        prev_x = new_x
        gradient = derivative_function(prev_x)
        new_x = prev_x - step_multiplier*gradient
        difference = abs(prev_x-new_x)
        xlist.append(prev_x)
        dxlist.append(derivative_function(prev_x))
        if (difference < precision):
            break
    return new_x,gradient,xlist,dxlist

# function call
minima,gradient,xlist,dxlist = gradient_descent(dh,-.2)

print(len(xlist))
print(len(dxlist))
```

```python
# Plot Section

plt.figure(figsize=[15,5])
plt.subplot(1,2,1)

plt.title("Cost Function",fontsize=17)
plt.xlabel("X",fontsize=16)
plt.ylabel("h(x)",fontsize=16)
           
plt.plot(x_3,h(x_3))
plt.scatter(xlist,h(np.array(xlist)),s=100,color='red',alpha=.5)
plt.xlim(-2, 4)
plt.ylim(-10, 10)
plt.grid()



plt.subplot(1,2,2)
plt.title("Slope of Cost Function",fontsize=17)
plt.xlabel("X",fontsize=15)
plt.ylabel("dh(x)",fontsize=15)
plt.plot(x_3,dh(x_3),linewidth = 3, alpha = .8)
plt.scatter(xlist,dxlist,color="red",s=100,alpha=.6)
plt.xlim(-2, 2)
plt.ylim(-5, 5)
plt.grid()

plt.show()
```
