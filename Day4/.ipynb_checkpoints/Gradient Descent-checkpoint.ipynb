# Gradient Descent Algorithm and It's visualization

### Below example for function 
- $ f(x) = x^2 + x + 1 $ 

- $ f'(x) = 2x + 1 $

```python
# Definition of function and it's derivative

def f(x):
    return x**2 + x + 1
def df(x):
    return 2*x +1
```

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Points to make a plot
x_1 = np.linspace(start=-3.5,stop = 2.5,num = 100)
```

```python
# to show x,y axis
axis = np.linspace(-20,20,5)
#type(axis)
```

## Plot of Function and It's Derivative

```python
plt.figure(figsize=[15,6]) 
#subplot1
plt.subplot(1,2,1)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,f(x_1))
plt.xlabel("X", fontsize=16)
plt.ylabel("F(x)",fontsize =16)
plt.title("f(x)=x^2+x+1")
plt.xlim(-3.9,3.5)
plt.ylim(-12,24)
plt.grid(axis="both",which="major")

#subplot 2
plt.subplot(1,2,2)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,df(x_1))
plt.xlim(-3.9,3.5)
plt.ylim(-12,12)
plt.grid()
plt.show()
```

## Gradient Descent Algorithm

```python
#Gradient Descent Algorithm
# Gradient approaches 0 as the iterations increases
new_x = 3
prev_x = 0
step_multiplier = .1
for i in range(50):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - (step_multiplier * gradient)

print(f"Local Minima or New_x = {new_x}")
print(f"Cost at minima = {f(new_x)}")
print(f"Gradient at Local Minima = {gradient}")
```

## Visualization of Gradient Descent Algorithm

```python

new_x = 3
prev_x = 0
step_multiplier = .1
x_list = [new_x]
for i in range(50):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - (step_multiplier * gradient)
    x_list.append(new_x)
values = np.array(x_list)

# Plotting Arrangements
plt.figure(figsize=[20,5])

#subplot 1
plt.subplot(1,2,1)

plt.scatter(x_list,f(values),color="red",alpha=.6)

plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,f(x_1))
plt.xlabel("X", fontsize=16)
plt.ylabel("F(x)",fontsize =16)
plt.title("f(x)=x^2+x+1")
plt.xlim(-3.9,3.5)
plt.ylim(-2,15)
plt.grid()

#subplot 2
plt.subplot(1,2,2)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,df(x_1))
plt.xlim(-1,3)
plt.ylim(-8,8)
plt.scatter(x_list, df(values),color="red",alpha=.6)
plt.grid()
plt.show()


```
