# Find graph of a function and it's derivative


```python
def f(x):
    return x**2 + x + 1
def df(x):
    return 2*x +1

#change return statement with your suitable function
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.linspace(start=-3.5,stop = 2.5,num = 100)
```

```python
# to show x,y axis
axis = np.linspace(-20,20,10)
```

```python
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis

plt.plot(x_1,f(x_1))
plt.xlabel("X", fontsize=16)
plt.ylabel("F(x)",fontsize =16)
plt.title("f(x)=x^2+x+1")
plt.xlim(-3.9,3.5)
plt.ylim(-3,12)
plt.show()
```

## Use of SubPlot to create multiple plots in single line

```python
plt.figure(figsize=[15,5])
#subplot1
plt.subplot(1,2,1)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,f(x_1))
plt.xlabel("X", fontsize=16)
plt.ylabel("F(x)",fontsize =16)
plt.title("f(x)=x^2+x+1")
plt.xlim(-3.9,3.5)
plt.ylim(-3,12)

#subplot 2
plt.subplot(1,2,2)
plt.plot(axis*0,axis,color="black")   #y-axis
plt.plot(axis,axis*0,color="black")   #x-axis
plt.plot(x_1,df(x_1))
plt.show()
```
