## 3D visualization

$$ f(x,y) = \frac{1}{3^{-x^2-y^2}  + 1} $$

or 

minimise $$ f(x,y)= \frac{1}{r+1} $$ where $r$ = $3^{-x^2-y^2}$

```python
def f(x,y):
    r = 3 **( -(x**2) - (y**2)) 
    return 1/(r+1)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm  # cm is color map
```

```python
x_1 = np.linspace(start=-2,stop=2,num=200)
y_1 = np.linspace(start=-2,stop=2,num=200)
x_1,y_1 = np.meshgrid(x_1,y_1)
f_1 = f(x_1,y_1)


# print(f_1.shape)
```

```python
fig = plt.figure(figsize=[20,12])
ax=fig.gca(projection="3d")

ax.set_xlabel("X",fontsize=16,color='red')
ax.set_ylabel("Y",fontsize=16,color='red')
ax.set_zlabel("F(x,y)",fontsize=16,color='red')

ax.plot_surface(x_1,y_1,f_1,color="yellow",cmap=cm.cool,alpha = .5)
plt.show()
# plt.plot(x_1,y_1,f_1)
```
